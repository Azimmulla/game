```python
import tkinter as tk
from tkinter import messagebox

class TreasureHuntGame:
    def __init__(self, root):
        self.root = root
        self.root.title("Treasure Hunt Game")

        self.current_question_index = 0
        self.questions = [
            {
                "question": "I speak without a mouth and hear without ears. I have no body, but I come alive with the wind. What am I?",
                "options": ["A. River", "B. Echo", "C. Mountain", "D. Cloud"],
                "answer": "B"
            },
            {
                "question": "The more you take, the more you leave behind. What am I?",
                "options": ["A. Footsteps", "B. Money", "C. Time", "D. Memories"],
                "answer": "A"
            },
            {
                "question": "What has keys but can't open locks?",
                "options": ["A. Piano", "B. Door", "C. Computer", "D. Treasure Chest"],
                "answer": "A"
            }
            # Add more questions as needed
        ]

        self.create_widgets()

    def create_widgets(self):
        self.question_label = tk.Label(self.root, text="", font=("Helvetica", 14))
        self.question_label.pack(pady=20)

        self.var = tk.StringVar()

        for i in range(4):
            option_button = tk.Radiobutton(
                self.root,
                text="",
                variable=self.var,
                value="",
                font=("Helvetica", 12)
            )
            option_button.pack(pady=5)
            setattr(self, f"option{i+1}_button", option_button)

        next_button = tk.Button(self.root, text="Next", command=self.check_answer)
        next_button.pack(pady=20)

        self.load_question()

    def load_question(self):
        if self.current_question_index < len(self.questions):
            question_data = self.questions[self.current_question_index]
            self.question_label.config(text=question_data["question"])

            options = question_data["options"]
            for i in range(4):
                option_button = getattr(self, f"option{i+1}_button")
                option_button.config(text=options[i], value=options[i][0])

        else:
            messagebox.showinfo("Game Over", "Congratulations! You've completed the treasure hunt.")

    def check_answer(self):
        if self.current_question_index < len(self.questions):
            selected_answer = self.var.get()
            correct_answer = self.questions[self.current_question_index]["answer"]

            if selected_answer == correct_answer:
                messagebox.showinfo("Correct", "Well done! You've found the next clue.")
                self.current_question_index += 1
                self.load_question()
            else:
                messagebox.showerror("Incorrect", "Oops! That's not the right answer. Try again.")

if __name__ == "__main__":
    root = tk.Tk()
    game = TreasureHuntGame(root)
    root.mainloop()
```
